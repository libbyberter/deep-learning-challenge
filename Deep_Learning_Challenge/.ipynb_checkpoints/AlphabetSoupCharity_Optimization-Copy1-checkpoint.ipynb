## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python
#  Import and read the charity_data.csv.
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['EIN', 'NAME'])
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

# Optimization 1 - increase number of bins in Application Type and Classification

```python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
# Optimization 1 - changed cutoff from <528 to <50
application_types_to_replace = ['T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17']

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
# df = df.value_counts().rename_axis('unique_values').reset_index(name='counts')
df = application_df['CLASSIFICATION'].value_counts().rename_axis('unique_values').reset_index(name='counts')
print(df[df['counts']>1])
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# Optimization 1 - changed cutoff from <1823 to <50
classifications_to_replace = []

for ind in df.index:
    if (df['counts'][ind])<50:
        classifications_to_replace.append(df['unique_values'][ind]) 
        
print(classifications_to_replace)
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

application_df_numeric = pd.get_dummies(application_df, 
                            columns=['APPLICATION_TYPE','AFFILIATION','CLASSIFICATION',
                                     'USE_CASE','ORGANIZATION','INCOME_AMT',
                                     'SPECIAL_CONSIDERATIONS'], dtype=int)

application_df_numeric.head()
```

```python
application_df_numeric.shape
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df_numeric['IS_SUCCESSFUL'].values
X = application_df_numeric.drop(columns='IS_SUCCESSFUL').values
print(X.shape)
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

##  Attempt 1 - Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.


nn_1 = tf.keras.models.Sequential()

# First hidden layer
nn_1.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=55))

# Second hidden layer
nn_1.add(tf.keras.layers.Dense(units=30, activation="relu"))

# Output layer
nn_1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_1.summary()
```

```python
# Compile the model
nn_1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_1.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Optimization 4 - use keras-tuner

```python
# Import the kerastuner library

import keras_tuner as kt
```

```python
#  Import and read the charity_data.csv.
application_df_4 = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df_4.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df_4 = application_df_4.drop(columns=['EIN', 'NAME'])

application_df_4 = application_df_4.drop(columns=
                                         ['STATUS','SPECIAL_CONSIDERATIONS'])

application_df_4.head()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df_4['ASK_AMT'].value_counts()
```

```python
df = application_df_4['ASK_AMT'].value_counts().rename_axis('unique_values').reset_index(name='counts')


df2 = df.sort_values('unique_values')
print(df2[df2['counts']>10])
```

```python
# Choose a cutoff value and create a list of ASK_AMT to be replaced
# use the variable name `ask_amt_to_replace`
# cutoff at 5000
ask_amt_to_replace = []

for ind in df2.index:
    if (df2['counts'][ind])<5000:
        ask_amt_to_replace.append(df2['unique_values'][ind]) 
        
# print(ask_amt_to_replace)
```

```python
# Replace in dataframe
for amt in ask_amt_to_replace:
    application_df_4['ASK_AMT'] = application_df_4['ASK_AMT'].replace(amt,"Other")
    
# Check to make sure binning was successful
application_df_4['ASK_AMT'].value_counts()

```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#cutoff at 50
application_types_to_replace = ['T12', 'T2', 'T25', 'T14', 'T29', 'T15', 'T17']

# Replace in dataframe
for app in application_types_to_replace:
    application_df_4['APPLICATION_TYPE'] = application_df_4['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df_4['APPLICATION_TYPE'].value_counts()
    
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
# cutoff at 5
classifications_to_replace = []

df = application_df_4['CLASSIFICATION'].value_counts().rename_axis('unique_values').reset_index(name='counts')


for ind in df.index:
    if (df['counts'][ind])<100:
        classifications_to_replace.append(df['unique_values'][ind]) 
        
print(classifications_to_replace)
```

```python
# Replace in dataframe
for cls in classifications_to_replace:
    application_df_4['CLASSIFICATION'] = application_df_4['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df_4['CLASSIFICATION'].value_counts()

application_df_4.shape
```

```python
# Convert categorical data to numeric with `pd.get_dummies`

application_df_4_numeric = pd.get_dummies(application_df_4, 
                             columns=['APPLICATION_TYPE','AFFILIATION','CLASSIFICATION',
                                     'USE_CASE','ORGANIZATION','INCOME_AMT','ASK_AMT'], dtype=int)

application_df_4_numeric.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df_4_numeric['IS_SUCCESSFUL'].values
X = application_df_4_numeric.drop(columns='IS_SUCCESSFUL').values
print(X.shape)
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn_4 = tf.keras.models.Sequential()

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','elu'])
    optimizer = hp.Choice('optimizer',['SGD','Adam'])
    
    # Allow kerastuner to decide number of neurons in first layer
    nn_4.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=100,
        step=4), activation=activation, input_dim=49))

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 5)):
        nn_4.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=100,
            step=4),
            activation=activation))
    
    nn_4.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    nn_4.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["accuracy"])
    
    return nn_4
```

```python
tuner = kt.BayesianOptimization(
    create_model,
    objective="val_accuracy",
    max_trials=10)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=60,validation_data=(X_test_scaled,y_test))
```
